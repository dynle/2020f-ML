```python
# https://www.statista.com/statistics/1102816/coronavirus-covid19-cases-number-us-americans-by-day/

import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("us_cases.xlsx",sheet_name='Data')
n=len(df["Unnamed: 2"])
y=df["Unnamed: 2"][4:n]
y=y.replace("-",0)
x=np.arange(4,n)
```

```python
trendpoly=np.poly1d(np.polyfit(x,y,10))
```

```python
xp=np.linspace(4,n)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,12))
plt.plot(x,y,label='data')
plt.plot(xp,trendpoly(xp),'--r',label='trend',linewidth=2.5)
plt.ylabel("Number of new cases",fontsize=20)
plt.legend(fontsize=20)
plt.show()
```

```python
x_new=np.linspace(0,n+20)

plt.figure(figsize=(12,12))
plt.plot(x,y,label='data')
plt.plot(x_new,trendpoly(x_new),'--r',label='predict',linewidth=2.5)
plt.ylabel("Number of new cases",fontsize=20)
plt.legend(fontsize=20)
plt.show()
```
